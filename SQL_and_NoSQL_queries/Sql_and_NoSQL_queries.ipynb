This notebook contains examples of queries relating to key business issues

# OLAP 

### request regarding income analysis

```python
-- Total monthly income 
SELECT 
    dt.month,
    dt.year,
    SUM(ft.transaction_amount) AS Total_income 
FROM fact_transaction ft
JOIN dim_time df ON ft.time_key = dt.time_key 
WHERE dt.year = 2024 
GROUP BY dt.year, dt.month
ORDER BY dt.month

-- Total revenue by country
SELECT 
  dg.country,
  SUM(ft.transaction_amount) AS total_revenue
FROM fact_transaction ft
JOIN dim_geography dg ON ft.geography_key = dg.geography_key
GROUP BY dg.country
ORDER BY total_revenue DESC

-- With the materialized view 
SELECT 
  date,
  total_revenue,
  transaction_count
FROM summary_daily_revenue
WHERE date >= '2024-01-01'
ORDER BY date
```

### request regarding fraud detection 

```python
-- by country 
SELECT 
  country,
  fraud_count,
  fraud_rate
FROM summary_fraud_by_country
ORDER BY fraud_rate DESC

-- By type fraud
SELECT 
  fraud_type,
  COUNT(*) AS total_frauds,
  SUM(fraud_amount) AS total_amount
FROM fact_fraud
GROUP BY fraud_type
ORDER BY total_frauds DESC
```
